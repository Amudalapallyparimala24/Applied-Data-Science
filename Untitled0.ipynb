```python
import pandas as pd

# Function to read and transform CO2 data
def read_and_transform_co2_data(filename):
    df = pd.read_csv(filename)
    df_transposed = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], var_name="Year", value_name="CO2 Emissions")
    df_pivoted = df_transposed.pivot_table(index="Year", columns="Country Name", values="CO2 Emissions").reset_index()
    df_transposed['Year'] = pd.to_datetime(df_transposed['Year'], format='%Y')
    return df_transposed, df_pivoted

# Function to read and transform GDP data
def read_and_transform_gdp_data(filename):
    df = pd.read_csv(filename)
    df_transposed = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], var_name="Year", value_name="GDP")
    df_pivoted = df_transposed.pivot(index='Year', columns='Country Name', values='GDP')
    df_transposed = df_transposed.drop(columns=["Country Code", "Indicator Name", "Indicator Code"])
    return df_transposed, df_pivoted

# Function to read and transform Population Growth data
def read_and_transform_population_data(filename):
    df = pd.read_csv(filename)
    df_years = df.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='Population')
    df_countries = df.set_index('Country Code').T.reset_index()
    df_countries.columns.name = None
    df_countries = df_countries.rename(columns={'index': 'Year'})
    return df_years, df_countries

# Function to read and transform Power Consumption data
def read_and_transform_power_consumption_data(filename):
    df = pd.read_csv(filename, encoding='ISO-8859-1')
    df_years = df.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='Power Consumption')
    df_countries = df.set_index('Country Name').T.reset_index()
    df_countries.columns.name = None
    df_countries = df_countries.rename(columns={'index': 'Year'})
    return df_years, df_countries

# Function to clean the dataframes
def clean_dataframes(*dataframes):
    cleaned_dataframes = []
    for df in dataframes:
        cleaned_df = df.dropna()  # Remove null value rows
        cleaned_df = cleaned_df.drop_duplicates()  # Remove duplicates
        for column in cleaned_df.columns:
            if pd.api.types.is_numeric_dtype(cleaned_df[column]):
                median = cleaned_df[column].median()
                std = cleaned_df[column].std()
                outliers = (cleaned_df[column] - median).abs() > 2 * std
                cleaned_df.loc[outliers, column] = median  # Replace outliers with median value
        cleaned_dataframes.append(cleaned_df)
    return cleaned_dataframes

# Load and transform the data
co2_transposed_df, co2_pivoted_df = read_and_transform_co2_data('CO2.csv')
gdp_transposed_df, gdp_pivoted_df = read_and_transform_gdp_data('GDP.csv')
population_years_df, population_countries_df = read_and_transform_population_data('Population Growth.csv')
power_years_df, power_countries_df = read_and_transform_power_consumption_data('Power Consumption.csv')

# Clean the dataframes
cleaned_co2_transposed, cleaned_co2_pivoted, cleaned_gdp_transposed, cleaned_gdp_pivoted, cleaned_population_years, cleaned_population_countries, cleaned_power_years, cleaned_power_countries = clean_dataframes(co2_transposed_df, co2_pivoted_df, gdp_transposed_df, gdp_pivoted_df, population_years_df, population_countries_df, power_years_df, power_countries_df)
```
