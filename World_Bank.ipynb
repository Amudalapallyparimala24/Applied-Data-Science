```python
import pandas as pd

def ingest_and_clean_data(file_path, countries, columns_to_drop):
    # Read the data into a dataframe
    df = pd.read_csv(file_path)

    # Filter the data for specified countries
    filtered_data = df[df['Country Name'].isin(countries)]

    # Remove specified columns
    cleaned_data = filtered_data.drop(columns=columns_to_drop)

    # Transpose the data
    cleaned_data_transposed = cleaned_data.T.reset_index()

    # Set the first row as column headers
    cleaned_data_transposed.columns = cleaned_data_transposed.iloc[0]
    cleaned_data_transposed = cleaned_data_transposed[1:]

    return cleaned_data, cleaned_data_transposed

# Define the countries and columns to drop
brics_countries = ['Brazil', 'Russian Federation', 'India', 'China', 'South Africa']
columns_to_drop = [str(year) for year in range(1960, 1990)] + ['2021', '2022']

# Ingest and clean CO2 data
brics_emissions, brics_emissions_transposed = ingest_and_clean_data("CO2.csv", brics_countries, columns_to_drop)

# Ingest and clean GDP data
brics_gdp, brics_gdp_transposed = ingest_and_clean_data("GDP.csv", brics_countries, columns_to_drop)

# Ingest and clean Forest Area data
brics_forest_area, brics_forest_area_transposed = ingest_and_clean_data("Forest Area.csv", brics_countries, columns_to_drop)

# Ingest and clean Urban Population data
brics_urban_population, brics_urban_population_transposed = ingest_and_clean_data("Urban Population.csv", brics_countries, columns_to_drop)

```
