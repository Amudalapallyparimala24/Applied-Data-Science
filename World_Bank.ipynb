```python
import pandas as pd

def read_world_bank_data(filename):
    # Read the data into a dataframe
    df = pd.read_csv(filename)

    # Transpose the dataframe to have years as columns
    df_transposed = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], var_name="Year", value_name="Value")

    # Pivot the transposed dataframe to have countries as columns
    df_pivoted = df_transposed.pivot_table(index=["Year"], columns="Country Name", values="Value").reset_index()

    return df_transposed, df_pivoted

# Call the function with the filename
transposed_df, pivoted_df = read_world_bank_data('CO2.csv')

# Clean the transposed dataframe if necessary
# For example, you can convert the 'Year' column to datetime
transposed_df['Year'] = pd.to_datetime(transposed_df['Year'], format='%Y')
```
