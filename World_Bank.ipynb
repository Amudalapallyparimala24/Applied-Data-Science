```python
import pandas as pd

def read_world_bank_data(filename):
    # Read the data into a dataframe
    df = pd.read_csv(filename)

    # Transpose the dataframe to have years as columns
    df_transposed = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], var_name="Year", value_name="Co2 Emisons")

    # Pivot the transposed dataframe to have countries as columns
    df_pivoted = df_transposed.pivot_table(index=["Year"], columns="Country Name", values="Co2 Emisons").reset_index()

    return df_transposed, df_pivoted

# Call the function with the filename
co2_transposed_df, co2_pivoted_df = read_world_bank_data('CO2.csv')

# Clean the transposed dataframe if necessary
# For example, you can convert the 'Year' column to datetime
co2_transposed_df['Year'] = pd.to_datetime(co2_transposed_df['Year'], format='%Y')
```

```python
import pandas as pd

def read_world_bank_data(filename):
    # Read the data into a dataframe
    df = pd.read_csv(filename)

    # Transpose the dataframe to have years as columns
    df_transposed = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], var_name="Year", value_name="GDP")

    # Pivot the transposed dataframe to have countries as columns
    df_pivoted = df_transposed.pivot(index='Year', columns='Country Name', values='GDP')

    # Clean the transposed dataframe
    df_transposed = df_transposed.drop(columns=["Country Code", "Indicator Name", "Indicator Code"])

    return df_transposed, df_pivoted

# Call the function with the filename
gdp_transposed_df, gdp_pivoted_df = read_world_bank_data('GDP.csv')

# Display the resulting dataframes

gdp_transposed_df.head()
```

```python
import pandas as pd

def read_and_transform_data(filename):
    # Read the data into a pandas dataframe
    df = pd.read_csv(filename)

    # Transpose the dataframe to have years as columns
    df_years = df.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='Population')

    # Transpose the dataframe to have countries as columns
    df_countries = df.set_index([ 'Country Code']).T.reset_index()

    return df_years, df_countries

# Call the function with the filename
filename = 'Population Growth.csv'
df_years, df_countries = read_and_transform_data(filename)

# Clean the transposed dataframe
df_countries.columns.name = None  # Remove the name of the columns index
df_countries = df_countries.rename(columns={'index': 'Year'})

df_years.head()
```
